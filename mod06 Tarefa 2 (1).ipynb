####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar os dados do arquivo CSV
dados_sinasc = pd.read_csv('SINASC_RO_2019.csv')

# Criar o boxplot da idade da mãe
plt.boxplot(dados_sinasc['IDADEMAE'])

# Configurar o título e os rótulos dos eixos
plt.title('Boxplot da Idade da Mãe')
plt.xlabel('Idade')
plt.ylabel('Idade da Mãe')

# Exibir o gráfico
plt.show()
```

####  3) Faça um histograma da idade da mãe

```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar os dados do arquivo CSV
dados_sinasc = pd.read_csv('SINASC_RO_2019.csv')

# Criar o histograma da idade da mãe
plt.hist(dados_sinasc['IDADEMAE'], bins=10)  # Ajuste o número de bins conforme necessário

# Configurar o título e os rótulos dos eixos
plt.title('Histograma da Idade da Mãe')
plt.xlabel('Idade')
plt.ylabel('Frequência')

# Exibir o gráfico
plt.show()
```
